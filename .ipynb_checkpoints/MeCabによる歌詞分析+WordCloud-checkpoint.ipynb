### 「名もなき詩」の歌詞を分解し、WordCloudで特徴を図示する

```python
txt = "NamonakiUta.txt"
with open(txt) as f:
    lyric = f.read()
print(lyric[:15])
```

```python
import MeCab
from matplotlib import pyplot as plt
from wordcloud import WordCloud
import numpy as np
from PIL import Image

path = "-d /usr/local/lib/mecab/dic/mecab-ipadic-neologd"
tagger = MeCab.Tagger(path)

# 形態素分解したnodeを取得する
node = tagger.parseToNode(lyric)

word_list = []

# 名詞、形容詞、動詞のみを取得する
while node:
    word_type = node.feature.split(',')[0]
    if word_type == '名詞' or word_type == '形容詞' or word_type =="動詞":
        word_list.append(node.surface)
    node = node.next

# リストを文字列に変換
word_chain = ' '.join(word_list)

# maskする画像を指定可能
mask_array = np.array(Image.open("chara_10_school-22.jpg"))

# ワードクラウド作成
W = WordCloud(width=1920, height=1080, 
              background_color='white', 
              font_path='/Library/Fonts/ヒラギノ角ゴシック W4.ttc', 
              max_words=500, 
              regexp=r"[\w']+",
              stopwords=(),
              colormap="rainbow",
              mask=mask_array
             ).generate(word_chain)

plt.imshow(W)
plt.axis('off')
plt.show()
```

## Word2vecで全ての歌詞をベクトル分析する

```python
import openpyxl
wb = openpyxl.load_workbook("lovelive_songs_data.xlsx")
niji_sheet = wb["nijigasaki_songs_data"]
aq_sheet = wb["uranohoshi_songs_data"]
myu_sheet = wb["otonokizaka_songs_data"]
lie_sheet = wb["yuigaoka_songs_data"]

import re

all_kashi2 = ""
data_list=[niji_sheet, aq_sheet, myu_sheet, lie_sheet]

for sheet in data_list:
    lyrics = [cell.value for cell in sheet["F"]]
    for i in range(1,len(lyrics)):
        lyric = re.sub(r"\s", " ", lyrics[i])  
        lyrics[i] = re.sub(r"[\(, \), \「, \」]", " ", lyric)
        all_kashi2 += lyrics[i]
```

```python
#全ての歌詞をtxtに書き込む

with open("全ての歌詞.txt", encoding="UTF-8", mode="w") as f:
    f.write(all_kashi2)
```

```python
import MeCab
import codecs

#Mecab で分かち書きをして, その結果を出力する
tagger = MeCab.Tagger('-Owakati -d /usr/local/lib/mecab/dic/mecab-ipadic-neologd')

fi = codecs.open('全ての歌詞.txt', 'r', 'utf8')
fo = codecs.open('全ての歌詞_wakati.txt', 'w', 'utf8')

line = fi.readline()

while line:
    result = tagger.parse("　"+line)
    #print(result)
    fo.write(result[1:])
    line = fi.readline()

fi.close()
fo.close()
```

```python
from gensim.models import word2vec

sentences = word2vec.LineSentence('全ての歌詞_wakati.txt')
model = word2vec.Word2Vec(sentences,
                          sg=1,
                          size=300,
                          min_count=5,
                          window=10,
                          hs=1,
                          negative=0,
                          iter=50)
model.save('./lovelive_word2vec')

model = word2vec.Word2Vec.load("./lovelive_word2vec")
word_vectors = model.wv.index2word
print(word_vectors)
```

```python
similar_words = model.wv.most_similar(positive=["あなたと"], topn=17)
print(*[" ".join([v, str("{:.2f}".format(s))]) for v, s in similar_words], sep="\n")
```
